```python
top_10 = {}
for index, row in df.iterrows():
    if row['rank'] <= 10:
        tba = {row['rank']:row['owner_ship']}
        top_10.update(tba)
    else:
        break
```

```python
public_counter = 0
private_counter = 0
for key in top_10:
    if top_10[key] == "Public/Government":
        public_counter = public_counter + 1
    elif top_10[key] == "Private":
        private_counter = private_counter + 1  
```

```python
x = ["Public","Private"]
h = [public_counter,private_counter]
c = ["pink","yellow"]
plt.bar(x,h,width=0.4, color=c,edgecolor="black")
plt.xlabel("Sector")
plt.ylabel("Number of Universities")
plt.title("Top 10 Universities")
plt.show()
```
